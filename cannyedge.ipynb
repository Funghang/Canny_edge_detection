```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#load image in grayscale
img = cv2.imread("D:\\picfolder\\fung.jpg")
img = cv2.resize(img,(90,150))
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

canny = cv2.Canny(img_gray,100,190)

plt.subplot(1,3,1),plt.imshow(img),plt.title("original_image")
plt.subplot(1,3,2),plt.imshow(img_gray),plt.title("grayscale")
plt.subplot(1,3,3),plt.imshow(canny, cmap="gray"),plt.title("canny")
cv2.imshow("original==",img)
cv2.imshow("gray====",img_gray)
cv2.imshow("canny",canny)
plt.savefig("canny.png")
#cv2.waitkey(0)
#cv2.destroyAllWindows()
```
